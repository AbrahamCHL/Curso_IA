# MatPlotLib 

## Dibujar un gráfico de línea

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.plot([1,2,3,4], [5,8,1,2])
plt.ylabel('Números')
plt.show()
```

## Mutiples funciones un un gráfico

```python
plt.plot([1,2,3,4], [5,8,1,2])
plt.plot([2,3,4,5], [5,8,1,2])
plt.show()
```

## Guardar en un archivo

```python
plt.plot([1,2,3,4], [5,8,1,2])
plt.plot([2,3,4,5], [5,8,1,2])
plt.savefig('C:\\temp\\mi_grafico.png', format='png')
```

## Se pueden ajustar los ejes

```python
axes = plt.axes()
axes.set_xlim([-1, 5])
axes.set_ylim([-1, 10])
plt.plot([1,2,3,4], [5,8,1,2])
plt.plot([2,3,4,5], [5,8,1,2])
plt.show()
```

## Añadir parrilla

```python
axes = plt.axes()
axes.set_xlim([-1, 5])
axes.set_ylim([-1, 10])
axes.grid()
plt.plot([1,2,3,4], [5,8,1,2])
plt.plot([2,3,4,5], [5,8,1,2])
plt.show()
```

## Cambiar los tipos de línea y su color

```python
axes = plt.axes()
axes.set_xlim([-1, 5])
axes.set_ylim([-1, 10])
axes.grid()
plt.plot([1,2,3,4], [5,8,1,2], 'bs-')
plt.plot([2,3,4,5], [5,8,1,2], 'r:')
plt.show()
```

## Poniendo títulos en los ejes y leyendas

```python
axes = plt.axes()
axes.set_xlim([-1, 5])
axes.set_ylim([-1, 10])
axes.grid()
plt.xlabel('Cantidad')
plt.ylabel('Probabilidad')
plt.plot([1,2,3,4], [5,8,1,2], 'bs-')
plt.plot([2,3,4,5], [5,8,1,2], 'r:')
plt.legend(['Azules', 'Rojos'], loc=2)
plt.show()
```

## Gráfico de torta

```python
values = [12, 30, 4, 32]
colors = ['r', 'g', 'b', 'c']
explode = [0, 0.2, 0.2, 0]
labels = ['Rusia', 'Argentina', 'Italia', 'Francia']
plt.pie(values, colors= colors, labels=labels, explode = explode)
plt.title('Cantidad de programadores')
plt.show()
```

## Gráfico de barras

```python
values = [12, 55, 4, 32, 14]
colors = ['r', 'g', 'b', 'c', 'm']
plt.bar(range(0,5), values, color= colors)
plt.show()
```
