```python
import pandas as pd

ratings = pd.read_csv('/home/supay/datos/rating.csv')
ratings = ratings[['user_id','anime_id','rating']]
movies = pd.read_csv('/home/supay/datos/anime.csv')
movies = movies[['anime_id','name']]

ratings = ratings.head(2000000)
ratings = pd.merge(movies, ratings)

dif = ratings.anime_id.unique()
print(len(dif))

#ratings.tail(100)
```

```python
animesRatings = ratings.pivot_table(index=['user_id'],columns=['name'],values='rating')

animesRatings
#print(animesRatings.keys())
```

```python
animeRatings = animesRatings['Ansatsu Kyoushitsu (TV) 2nd Season']
```

```python
similarMovies = animesRatings.corrwith(animeRatings)
similarMovies = similarMovies.dropna()

df = pd.DataFrame(similarMovies)
df.head(10)
```
