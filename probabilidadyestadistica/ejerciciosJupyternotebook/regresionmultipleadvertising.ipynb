```python
%matplotlib inline
import pandas as pd
import statsmodels.api as sm

df = pd.read_csv('./JupyterNotebooks/datos/Advertising.csv')
df.head()
```

```python
X = df[['TV', 'radio', 'newspaper']]
y = df['sales']

#Agregamos la intersecci√≥n en Y
X = sm.add_constant(X)
est = sm.OLS(y, X).fit()
est.summary()
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
X = df[['TV', 'radio', 'newspaper']]

#as_matrix() convierte de DataFrame a Numpy array
#Si queremos obtener el precio estimado, hay que cometar esta l√≠nea
X[['TV', 'radio', 'newspaper']] = scale.fit_transform(X[['TV', 'radio', 'newspaper']].values)
est = sm.OLS(y, X).fit()
est.summary()
```
