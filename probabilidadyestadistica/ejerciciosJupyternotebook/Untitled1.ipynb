```python
import pandas as pd

books_ratings = pd.read_csv('/home/supay/Descargas/Book_Ratings.csv',sep = ';', encoding="ISO-8859-1")

books = pd.read_csv('/home/supay/Descargas/Books.csv',sep = ';', encoding="ISO-8859-1",error_bad_lines=False)
books = books[['ISBN','Book-Title']]

books_ratings = books_ratings.head(9000)

ratings = pd.merge(books, books_ratings)
```

```python
userRatings = ratings.pivot_table(index=['User-ID'],columns=['ISBN'],values='Book-Rating')

userRatings.head()
```

```python
corrMatrix = userRatings.corr(method='pearson', min_periods=100)
corrMatrix.head()
```
